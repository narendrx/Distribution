Success Probability (p): 1/6 (since there are 6 sides on a die)
Number of Trials (n): 30
Number of Successes (x): 5.

2. Generate 100 values in Excel using the continuous uniform distribution RAND() and plot a histogram. Describe the shape of the distribution.
- If you generate 100 values using =RAND():
Shape: The histogram should look roughly rectangular or flat.
Description: Because every value between 0 and 1 has an equal probability of being selected, the distribution is Uniform. With only 100 values, it may look slightly jagged, but as you increase the sample size, it becomes perfectly flat.

3. A dataset has a mean of 50 and a standard deviation of 5. What percentage of values lie between 45 and 55 if the data follows a normal distribution?
- Mean (\mu): 50
Standard Deviation (\sigma): 5
Range: 45 to 55.
Notice that 45 is exactly one standard deviation below the mean (50 - 5) and 55 is one above (50 + 5). According to the Empirical Rule (68-95-99.7 rule), approximately 68% of values lie within one standard deviation of the mean.

4. What is the concept of standardization (z-score), and why is it important in data analysis? Explain the formula and how standardization transforms a dataset.
- Concept: Standardization converts data from different scales into a common scale with a mean of 0 and a standard deviation of 1.
- Importance: It allows you to compare data points from different datasets (e.g., comparing a math score to a history score) and identifies outliers.
- Transformation: It shifts the distribution to center on 0 and "squashes" or "stretches" it so the spread is measured in units of standard deviations.

5. What is Kurtosis and their type?
- Kurtosis measures the "tailedness" of a distribution—how much data is in the peaks versus the tails.
Mesokurtic: Normal distribution (Kurtosis ≈ 3 or Excess Kurtosis = 0).
Leptokurtic: Sharp peak and heavy tails (Positive Excess Kurtosis).
Platykurtic: Flat peak and thin tails (Negative Excess Kurtosis).

6. Explain why the uniform distribution is a good model for the outcome of rolling a fair die.
- A uniform distribution is the best model for a fair die because every outcome is equally likely. Since each number (1 through 6) has exactly a 1/6 chance of occurring, the probability mass function is constant across all possible results.

7. Use Excel to compute the probability of getting at least 8 successes in 15 trials with success probability 0.5.
- To find the probability of at least 8 successes in 15 trials (p = 0.5):
In Excel, BINOM.DIST with "TRUE" calculates the probability of x or fewer. To get "8 or more," we subtract the probability of "7 or fewer" from 1.
Excel Formula:
=1 - BINOM.DIST(7, 15, 0.5, TRUE)
Result: Approximately 0.50.

8. How does log transformation help in stabilizing variance and making data more normally distributed?
- Log transformation helps by:
Compressing Scale: It pulls in extreme outliers on the right side of a distribution, which reduces right-skewness to make data more "Normal."
Stabilizing Variance: It is useful when the variance increases proportionally with the mean (heteroscedasticity). By taking the log, the variance becomes more constant (homoscedastic) across the dataset.
